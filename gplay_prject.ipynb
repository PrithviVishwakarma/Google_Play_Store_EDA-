```python
def cleansize(col):
    if 'k' in col:
        col=col.replace('k','')
        return float(col)/1024

    elif 'M' in col:
        col=col.replace('M','')
        return float(col)

    elif 'Varies with device' in col:
        return np.nan

    else:
        return float(col)
```

```python
df['Size'] = df['Size'].map(cleansize)
```

```python
df.info()
```

```python
Size_mean = df['Size'].mean()
```

```python
df['Size'].fillna(Size_mean, inplace=True)
```
